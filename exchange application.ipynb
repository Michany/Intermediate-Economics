```python
import exchange
import numpy as np
import matplotlib.pyplot as plt

population=20
rich = [];poor=[]
for i in range(population):
    rich.append(exchange.Person(round(np.random.normal(3, 1),2), round(np.random.normal(9, 1),2), 'x**1.2*y'))
    # idea 1:;y^1.5
    # idea 2:xy
    poor.append(exchange.Person(round(np.random.normal(6, 1),2), round(np.random.normal(3, 1),2), 'x*y**1.5'))

poor_before=[[],[]];poor_after=[[],[]]
rich_before=[[],[]];rich_after=[[],[]]
for i in poor:poor_before[0].append(i.status()[0]);poor_before[1].append(i.status()[1])
for i in rich:rich_before[0].append(i.status()[0]);rich_before[1].append(i.status()[1])

price=list()
for i in range(population):
    for j in range(population):
        #print('exchanging Person %d and Person %d ...' % (i, j))
        exchange.change(rich[i], poor[j])
        price.append(1/rich[i].relative_price_2[poor[j]])
for r in range(100):
    # 
    rand = np.random.randint(low=0,high=population,size=6)
    exchange.change(rich[rand[0]], poor[rand[1]])
    exchange.change(rich[rand[2]], rich[rand[3]])
    exchange.change(poor[rand[4]], poor[rand[5]])
    price.append(1/rich[rand[0]].relative_price_2[poor[rand[1]]])
    price.append(1/rich[rand[2]].relative_price_2[poor[rand[3]]])
    price.append(1/rich[rand[4]].relative_price_2[poor[rand[5]]])

for i in poor:poor_after[0].append(i.status()[0]);poor_after[1].append(i.status()[1])
for i in rich:rich_after[0].append(i.status()[0]);rich_after[1].append(i.status()[1])

```

```python
plt.figure(figsize=(9,7))
plt.title('Scatter of people\'s status')
plt.xlabel('x (Stand for health)', fontsize=12)
plt.ylabel('y (Stand for money)', fontsize=12)

plt.scatter(poor_before[0],poor_before[1],marker = 'x',s = 55,label='poor_before', edgecolors= 'grey')
plt.scatter(poor_after[0],poor_after[1],marker = 'x',s = 55,label='poor_after', edgecolors= 'grey')

plt.scatter(rich_before[0],rich_before[1],label='rich_before',s = 40, edgecolors= 'grey')
plt.scatter(rich_after[0],rich_after[1],label='rich_after',s = 40, edgecolors= 'grey')

plt.grid(True)
plt.legend()

plt.savefig('test2png4.png', dpi=144)
```
